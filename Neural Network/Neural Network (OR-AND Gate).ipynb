```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# input data
X = np.array([[0,0],[0,1],[1,0],[1,1]])

# output data
y = np.array([0,0,0,1])

# weights
w = np.array([0.0,0.0])
# bias
b = 0.0

# learning rate
learning_rate = 0.1

# activation function
def activation(x):
    if x > 0:
        return 1
    else:
        return 0
    
# prediction function
def predict(X):
    z = np.dot(X,w) + b
    a = activation(z)
    return a

# training function
def train(X,y,epochs, b):
    for i in range(epochs):
        for j in range(len(X)):
            a = predict(X[j])
            error = y[j] - a
            w[0] = w[0] + learning_rate * error * X[j][0]
            w[1] = w[1] + learning_rate * error * X[j][1]
            b = b + learning_rate * error
    print(w,b)
    
# training
train(X,y,20, b)
 
# prediction
for i in range(len(X)):
    print("X: {} y: {} prediction: {}".format(X[i],y[i],predict(X[i])))
```

# OR GATE


```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1]) # output data
w = np.array([0.0,0.0]) # weights
b = 0.0 # bias
learning_rate = 0.1 # learning rate
def activation(x): # activation function
    if x > 0:
        return 1
    else:
        return 0
    
def predict(X): # prediction function
    z = np.dot(X,w) + b
    a = activation(z)
    return a

# training function
def train(X,y,epochs, b):
    for i in range(epochs):
        for j in range(len(X)):
            a = predict(X[j])
            error = y[j] - a
            w[0] = w[0] + learning_rate * error * X[j][0]
            w[1] = w[1] + learning_rate * error * X[j][1]
            b = b + learning_rate * error
    print(w,b)
    
# training
train(X,y,20, b)

# prediction
for i in range(len(X)):
    print("X: {} y: {} prediction: {}".format(X[i],y[i],predict(X[i])))
```
